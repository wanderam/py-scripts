```python
import random
```

### Situação número faltante

```python
# Criando uma sequência de 1 a 100
n_inicial = 1
n_final = 101
numeros = list(range(n_inicial, n_final))
# Embaralhando os números (inserindo aleatoriedade)
random.shuffle(numeros)
# Removendo um número aleatório
numeros.remove(random.choice(numeros))

print(f'A sequência de {n_inicial} a {n_final - 1} agora tem {len(numeros)} números!')
```

```python
"""
Somam-se todos os números que deveriam estar lá;
Subtrai-se a soma dos números que realmente estão lá;
O resultado é exatamente o número que foi removido.
"""

numero_faltante = sum(range(1, 101)) - sum(numeros)

print(f'O somatório de "numeros" é {sum(numeros)}')
print(f'O número faltante é {numero_faltante}.')
print(f'{sum(numeros)} + {numero_faltante} = {sum(numeros) + numero_faltante}')
```

### Situação número duplicado

```python
# Criando uma lista de 1-100 com um duplicado
numeros_ = list(range(1, 101))
duplicado = random.randint(1, 100)
numeros_.append(duplicado)
random.shuffle(numeros_)

print(f'A lista tem {len(numeros_)} números.')
```

```python
# Obtendo o somatório da sequência
somatorio = sum(range(1, 101))

# Resolvendo
numero_duplicado = sum(numeros_) - somatorio

print(f'O número duplicado é {numero_duplicado}.')
```
