```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro
```

```python
#Gerando dados aproximadamente normais

np.random.seed(42)
dados = np.random.normal(loc=100, scale=10, size=100)
df = pd.DataFrame(data=dados, columns=['Valores'])

df
```

```python
#Visualizando o histograma

plt.hist(df['Valores'], bins=30, color='gray')

plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma - Distribuição Aproximadamente Normal')

plt.show()
```

```python
#Aplicando o teste de Shapiro-Wilk

"""
Como funciona o teste de Shapiro-Wilk?

Hipótese nula (H₀): os dados seguem uma distribuição normal

Hipótese alternativa (H₁): os dados não seguem uma distribuição normal

Se o p-valor > 0.05, não rejeitamos H₀ → dados podem ser considerados normais
Se o p-valor ≤ 0.05, rejeitamos H₀ → dados não são normais
"""


stat, p_valor = shapiro(df['Valores'])

dash = '--'
n = 12
print(f'{dash * n} \nEstatística: {round(stat, 4)} \np_valor: {round(p_valor, 4)} \n{dash * n}')
```
